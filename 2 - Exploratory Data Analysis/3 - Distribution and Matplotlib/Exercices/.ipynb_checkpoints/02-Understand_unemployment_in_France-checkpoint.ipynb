* By doing a <a href="https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.pie.html" target="_blank">pie-chart</a>, show the share of unemployed, inactive and employed persons in France (from the variable ```ACTEU```).

Here you have some steps:
- Create a variable wich is the ```ACTEU```
- Convert this variable to qualitative values <a href="https://www.insee.fr/fr/statistiques/3555153#documentation" target"_blank">(refeer to the doc)</a>
- Then do the pie chart with these informations:

  * The percentage of each category

````python
dataset.head()
````

````python
acteu = dataset.ACTEU
acteu = acteu.replace({"1": "Active", "2": "Unemployed", "3": "Inactive", np.nan: "Unknow"})
cat_acteu = acteu.value_counts()
display(cat_acteu)
````

````python
import numpy as np
import matplotlib.pyplot as plt

explode = (0.1, 0.1, 0.2, 0.1)

plt.pie(cat_acteu.values, explode=explode , labels=cat_acteu.index, autopct='%1.1f%%', 
        radius=3, startangle=-270)
plt.legend(title = "ACTEU pie chart")
plt.show()
````

* Make the same graph on the variable ```ACTEU6``` which is more precise on the type of actives.

Remember:

* By ploting a <a href="https://pythonmatplotlibtips.blogspot.com/2018/11/normalized-stacked-barplot-number-percentage-python-matplotlib.html" target="_blank">stacked bar chart</a>, compare :

  * The ratio *Unemployed* / *Actives* 
  * The ratio *Unemployed* / *Actives + Inactives*

````python
unemployed = [cat_acteu.loc["Unemployed"] / cat_acteu.loc["Active"], 
              cat_acteu.loc["Unemployed"] / (cat_acteu.loc["Active"] + cat_acteu.loc["Inactive"])]

print("Ratio unemployed / actives:")
display(unemployed[0])
print("")
print("Ratio unemployed / actives + inactives:")
display(unemployed[1])
````

````python
employed = [1 - element for element in unemployed]
print("Active report for unemployed / actives:")
display(employed[0])
print("")
print("Active report for unemployed / actives + inactives:")
display(employed[1])
````

````python
names = ['Unemployed / Occuped actives', 'Unemployed / Actives + Inactives']
````

````python
# Create Barplot
f, ax = plt.subplots(figsize=(8,8))
plt.bar(names, unemployed)
plt.bar(names, employed, bottom=unemployed)

## On ajoute du texte avec la valeur ratios dans les barres pour plus de lisibilitÃ©
plt.text(names[0], 0.6, np.round(employed[0],2), ha="center", va="center")
plt.text(names[0], 0.05, np.round(unemployed[0],2), ha="center", va="center")

plt.text(names[1], 0.6, np.round(employed[1],2), ha="center", va="center")
plt.text(names[1], 0.025, np.round(unemployed[1],2),ha="center", va="center")

plt.title("Distribution of the unemployed in relation to the active population in the total population")
plt.show()
````

* Let's take a look at what pushes French people to change jobs. Through the ```CREACCP``` feature, create a horizontal bar chart that will allow you to know the main reasons why French people change jobs.

````python
creaccp = dataset.CREACCP
creaccp = creaccp.replace({"9": "Need or wants to move", 
                           "10": "Want to be freelance", 
                           "8": "Wants to reduce transport time", 
                           "5": "Need more suitable or flexible hourly rythm", 
                           "12": "Finds the work atmosphere bad, conflicting work relations", 
                           "11": "Wants to change job of job or activity sector", 
                           "4": "Wants to work more hours",
                           "1": "Risk of losing or gonna lose current job", 
                           "6": "Need to work with conditions that are less strenuous or more adapted to his or her health", 
                           "3": "Wants a CDI", 
                           "2": "Wants a more interesting job", 
                           "7": "Wants to increase salary",
                           "13": "Other reasons",
                           np.nan: "Unknow"})
creaccp = creaccp.value_counts().sort_values(ascending=True)
display(creaccp)
creaccp.drop(labels=["Unknow"], inplace=True)
````

* Calculate the sum of all features without ```Unknow```

````python
total_feature = creaccp.values.sum()
print("The sum of all feature without Unkow is: %d" %(total_feature))
````

- Calculate the repartition of each feature

````python
creaccp = creaccp.sort_values(ascending=False)
datas = {"Feature": creaccp.index, "Percentage": ((creaccp.values / total_feature) * 100)}

pd.DataFrame(datas)
````

````python
import seaborn as sns

sns.barplot(x="Percentage", y="Feature", data=datas, orient="h", color="b")

for i in range(len(datas.Percentage)):
    plt.text(datas.Percentage[i], datas.Feature[:-1][i],'{}%'.format(datas.Percentage[i]))

plt.title("Ranking of the reasons why French people want to change jobs")
plt.show()
````

* Let's end this exercise by looking at the distribution of desired hours for those looking for part-time work. To do this, use the HREC column. 

  * First try to simply create a histogram without the density function. 
  * To create the density function, you will need to use the following formula: 
      ```y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu))**2))```
     
     where *mu* is the mean of the distribution and *sigma* is the standard deviation.
