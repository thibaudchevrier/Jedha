## General operation of Matplotlib

### How to build a graph on Matplotlib

There are several things to consider when building a graph on Matplotlib. The diagram below should help you see more clearly.

![](https://matplotlib.org/3.1.1/_images/anatomy.png)


Mainly, you must remember the following hierarchical elements:

* The **Figure**: A *figure* is the entire frame within which your graphic(s) is contained. You will be able to define the size of your *figure*, the graphic number inside, etc.

* **Axes**: This corresponds to a graph within your *figure*

![](https://miro.medium.com/max/848/1*Fk80UWElsZI6f_poAv_7cQ.png)

* The **Axis**: Be careful not to confuse with the element above. The *Axis* correspond to the abscissa and ordinate of each of the graphs (or *Axes*).

* **Artist**: You may see this term in [Matplotlib](https://matplotlib.org/3.1.1/tutorials/introductory/usage.html#interactive-example) documentation. This refers to the set of elements in the graph (including *Figure*, *Axes*, *Axis*).

## The line plots

### What use ?

*Lines plots* are extremely useful for Time Series data types.

### How to?

In order to create a *line plot*, all you need to do is use `plt.plot()`.

````python
# Import librairies
import matplotlib.pyplot as plt
import numpy as np


# Create random sinusoidal data
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# LinePlot creation
plt.plot(t, s)
plt.show()
````

You can change the line color with the ```color```argument in ```plt.plot()```.

````python
# Import librairies
import matplotlib.pyplot as plt
import numpy as np


# Create random sinusoidal data
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Create lineplot
plt.plot(t, s, color="green")
plt.show()
````

## Histograms

### What use?

*Histograms* are very usefull to show the distribution. For example, the distribution of income according to age.

### How to?

The way to make a histogram is via ```plt.hist```.


````python
# Import librairies
import numpy as np
import matplotlib.pyplot as plt

# Create random data
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# Create histogram
plt.hist(x, 50, facecolor='g')
plt.show()
````

We specify in advance in the graph the number of bins: In this case 50.



### Some interesting parameters

#### Density

The histogram, as mentioned above, is often used to visualize a distribution. Rather than knowing raw values, we try instead to determine a probability distribution.

To do this, simply put the argument ```density = True```


````python
# Import librairies
import numpy as np
import matplotlib.pyplot as plt

# Create random data
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# Create histogram
plt.hist(x, 50, facecolor='g', density=True)
plt.show()
````

NB: If you don't see the difference between this graph and the graph above, look at the values on the Y-axis ;)



#### Creating a density curve

You may want to create a density curve. This allows you to no longer have *bins* (bars) but a continuous relationship between your variable and the probability you want to calculate.

To take the example of the distribution of income according to age, it is possible to want to know the income of a person whatever his age (25 years as 25,5391 years). This provides a more accurate picture of this income.

To do this, we will use the distribution formula for a normal law:

# $F(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{\frac{-(x-\mu)^2}{2\sigma^2}}$

In terms of code, it goes like this:

````python
# Import librairies
import numpy as np
import matplotlib.pyplot as plt

# Create random data
mu, sigma, bins = 100, 15, 50
x = mu + sigma * np.random.randn(10000)

# Create histogram
## We store in three different variables the results of plt.hist in n, bins and patches
## We'll use the variable bins which corresponds to our x in our formula

n, bins, patches = plt.hist(x, bins, facecolor='g', density=True)

# Density formula
## Like above
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

# Add a line plot
plt.plot(bins, y, color="r",linestyle="--")

plt.show()
````

## Pie charts

#### What use?

*Pie charts* are useful when you want to show the distribution of elements that are part of a whole. Typically, you want to know the distribution of men and women in France.

### How to?

This time you will have to use ```plt.pie``` to create your pie chart.


````python
# Import librairie
import matplotlib.pyplot as plt

# Pie chart where each of the elements will be arranged in a counter-clockwise direction
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True)
plt.show()
````

### Some usefull parameters
#### Explode

Sometimes it is useful to highlight a *cheese in the camembert* or, in more scientific terms, to highlight an element.

This can be done using the `explode' argument. This one takes a tuple of the size of the number of elements in the pie-chart. The value that you will put in each of the indices of the tuples will correspond to the distance that the element will be from the center of the pie chart.

Here's an example:


````python
# Import librairie
import matplotlib.pyplot as plt

# Pie chart where each of the elements will be arranged in a counter-clockwise direction
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0) # Here, we're only going to bring out the Hogs...

plt.pie(sizes, explode=explode,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
````

## Box plots

### What use?

*Box plots* are very useful when you want to know the distribution of a categorical variable. For example, you want to know the distribution of grades to a math test.

### How to?

To make a *boxplot*, you can use ```plt.boxplot```


````python
# Import the librairie
import matplotlib.pyplot as plt

# Create random data (for example grade in a math test)
math_test = np.random.randint(1, 20+1, 30)

# Create boxplot
plt.boxplot(math_test)
plt.show()
````

NB: If you run the code, you may not have the same *boxplot*, this is normal since the data is randomly generated via `np.random.randint()`



### **How to interpret a boxplot?**

A *boxplot* is useful to see where most of your data is. For example, in the graph above, we know that the majority of students had between 10 and 17.5. The big box represents in fact *the interquartile range* that is to say 50% of the dataset is in this box.

The orange line represents the *median*. In other words, half of the dataset had a score below 12 and the other half had a score above 12.

The rest of the *boxplot* represents minimum and maximum values.

It is also possible that the graphs give you *points outside the boxplot* which are this time *outliers*.

### Some useful parameters

#### Set an outlier limit

If you wish, you can use the `whis` argument, whose value you will insert will determine the values beyond which the elements in your sample will be considered outliers.

The calculation is done as follows:

```
Upper_bound = Q3 + whis*IQR
Lower_bound = Q1 - whis*IQR
```
Where IQR is the interquartile range (Q3-Q1)

````python
# Import the librairie
import matplotlib.pyplot as plt

# Create random data (for example grade in a math test)
math_test = np.random.randint(1, 20+1, 30)

# Add outliers
for i in [0, 8]:
  math_test[i] = 24

# Create boxplot
plt.boxplot(math_test,
            whis=0.9,
            labels=['Math Test'])
plt.show()
````

#### Add the mean to you boxplot

Viewing the mean on your graph is quite useful when you want to compare it to the median and see the difference. All you have to do is add the arguments:

* `showmeans=True`
* `meanline=True`

````python
# Import the librairie
import matplotlib.pyplot as plt

# Create random data (for example grade in a math test)
math_test = np.random.randint(1, 20+1, 30)

# Add outliers
for i in [0, 8]:
  math_test[i] = 24

# Create boxplot
plt.boxplot(math_test,
            whis=0.9,
            showmeans=True,
            meanline=True,
            labels=['Math Test'])
plt.show()
````

## Scatter plots

### What use?

*Scatter plots* are used to see the correlation between two variables. Frequently you will use it in Machine Learning

### How to?

You simply use ```plt.scatter``` to create a *scatter plot*.


````python
# Import librairies
import numpy as np
import matplotlib.pyplot as plt

# Create random data
x = np.arange(0.0, 50.0, 2.0)   # np.arange allows to create a list of numbers spaced by a defined step (here, we go from 2 to 2, 50 times)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0 # We create an approximately linear relationship in x and y

# Create scatter plot
plt.scatter(x, y)
plt.show()
````

### Some usefull parameters

#### Change the size of your points

Sometimes you may want to get bigger points for certain conditions. You can do it with the argument ```s```.


````python
# Import librairies
import numpy as np
import matplotlib.pyplot as plt

# Create random data
x = np.arange(0.0, 50.0, 2.0)   # np.arange allows to create a list of numbers spaced by a defined step (here, we go from 2 to 2, 50 times)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0 # We create an approximately linear relationship in x and y


# Add different sizes for each points
area = (15 * np.random.rand(50))**2  

# Create scatter plot
plt.scatter(x, y, s=area)
plt.show()
````

#### Change the color of points

You can also change the color of your points according to certain conditions via the argument ```c```


````python
# Import librairies
import numpy as np
import matplotlib.pyplot as plt

# Create random data
x = np.arange(0.0, 100, 2.0)   # np.arange allows to create a list of numbers spaced by a defined step (here, we go from 2 to 2, 50 times)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0 # We create an approximately linear relationship in x and y


# Add different heigh for each points
area = (15 * np.random.rand(50))**2  

# Add sequence of colors
c = ["b", "r"] * 25

# Create scatter plot
plt.scatter(x, y, s=area, c = c)
plt.show()
````

## Personnalize graphics

We've seen the main types of graphics you can do with Matplotlib but we haven't seen how to customize them.

It is important to be able to add titles, text and other such elements to your graphics. So that's the theme of this part.

### Add titles

To add a title, you just have to put ```plt.title()```


````python
# Import librairie
import matplotlib.pyplot as plt

# Create random data
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

# Create barplot
plt.bar(names, values)
plt.title("Voici un titre") # Here we build our title
plt.show()
````

### Add a legend

You can add legends via ```plt.legend()```


````python
# Import librairie
import matplotlib.pyplot as plt

# Create random data
names = ['group_a', 'group_b', 'group_c']
values_1 = [1, 10, 100]
values_2 = [7, 3, 10]

# Create Barplot
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1)
plt.title(names) # Create title
plt.legend(["values_1", "values_2"]) # Create legend
plt.show()
````

### Add labels to abscissa & ordinates

You can of course give names to the axes of the graph via ```plt.xlabel``` or ```plt.ylabels```


````python
# Import librairie
import matplotlib.pyplot as plt

# Create random data
names = ['group_a', 'group_b', 'group_c']
values_1 = [1, 10, 100]
values_2 = [7, 3, 10]

# Create Barplot
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1)
plt.title(names) # Create the title
plt.legend(["values_1", "values_2"]) # Create the legend
plt.xlabel("Abscissa name") # Create label for abscissa
plt.ylabel("Ordinate name") # Create label for ordinate
plt.show()
````

### Add text to your graph

Sometimes you need to add text inside your graphs to give the reader more visibility on the exact values. To do this, you can use ```plt.text```.

This method works as follows:

* You specify the coordinates of the text on the x-axis and y-axis.
* You specify the text you want to put

````python
# Import libraire
import matplotlib.pyplot as plt

# Create random data
names = ['group_a', 'group_b', 'group_c']
values_1 = [1, 10, 100]
values_2 = [7, 3, 10]

# Create Barplot
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1)
plt.title(names) # Create title
plt.legend(["values_1", "values_2"]) # Create legends
plt.xlabel("Abscissa name") # Create label for abscissa
plt.ylabel("Ordinate name") # Create label for ordinate
plt.text(names[0], 10, values_1[0]+values_2[0]) # Add text at the top of bars
plt.text(names[1], 20, values_1[1]+values_2[1])  # Add text at the top of bars
plt.text(names[2], 110, values_1[2]+values_2[2]) # Add text at the top of bars
plt.show()
````

### Resize your graph

Often you will need to enlarge the size of your graph, you can do this via ```plt.figure(figsize=(height, width))```


````python
# Import librairie
import matplotlib.pyplot as plt

# Create random data
names = ['group_a', 'group_b', 'group_c']
values_1 = [1, 10, 100]
values_2 = [7, 3, 10]

# Create barplot
plt.figure(figsize=(10,10)) # Resize the size of the figure
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1)
plt.title(names) # Create title
plt.legend(["values_1", "values_2"]) # Create legends
plt.xlabel("Abscissa name") # Create label for abscissa
plt.ylabel("Ordinate name") # Create label for ordinate
plt.text(names[0], 10, values_1[0]+values_2[0]) # Add text at the top of bars
plt.text(names[1], 20, values_1[1]+values_2[1])  # Add text at the top of bars
plt.text(names[2], 110, values_1[2]+values_2[2]) # Add text at the top of bars
plt.show()
````

### Do multiples graphs in only one figure

In particular, when you want to manage time series, you will probably want to add several graphs in the same figure.


````python
# Import librairie
import matplotlib.pyplot as plt
import numpy as np


# Create random sinusoidals data
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t) # Equation of the sin function
u = 1 + np.log(2 * t) # Equation of the log function

# Create the LinePlot
plt.plot(t, s, color="green", linestyle="--") # Create the sinusoidal function
plt.plot(t, u, color="blue") # Create the logarithmic function
plt.show()
````

### Have several graphics in the same output

Finally, you probably want to display several different graphs in the same *output*. For this, you can use ```plt.subplot()```

In the code above, each ``plt.subplot()`` contains a 3-digit number. It corresponds to *nrows*, *ncols*, *index*.
For example:

`plt.subplot(223)` means that there are **2 rows** in the whole figure, **2 columns** and that this subplot will be at the **3rd position**.

## Object oriented graphics for total flexibility

At the end of this course, you may see another way to create graphs: The OOP method.

You can set two `fig` & `ax` variables via `plt.subplot()`.

That way, you'll have:

````python
fig, ax = plt.subplots()
````

In the variable `fig` will be contained all elements related to your *figure* and in the variable *ax* will be contained all elements related to your graphic(s).

The operating system remains the same, however:

````python
# Import librairie
import matplotlib.pyplot as plt
import numpy as np

# Create data
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Create two variables
fig, ax = plt.subplots()

# Build the graph
ax.plot(t, s) # Add data in line

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks') # Here we set the title & lables for the abscissa & coordinates
plt.show()
````

Try to get used to both structures because you will be able to see both used interchangeably in different kernels.

This way of building graphs is more difficult to understand at the beginning but allows you to gain flexibility.
