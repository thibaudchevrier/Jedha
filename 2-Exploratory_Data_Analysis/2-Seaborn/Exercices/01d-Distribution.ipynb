Let's continue on <a href="https://full-stack-bigdata-datasets.s3.eu-west-3.amazonaws.com/Data+visualisation+et+collaboration/M03.D02/IBM_HR_ATTRITION.csv" target="_blank">```IBM_HR_ATTRITION.csv```</a>

1. Import usual librairies

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv("../../src/IBM_HR_ATTRITION.csv")
````

2. Using ```distplot()```, construct a graph that shows the distribution of revenues per month at IBM (```MonthlyIncome```).

````python
sns.displot(data = df["MonthlyIncome"], stat="density", kde=True)
````

3. We can see that the very high salaries bias our distribution, let's try to see the distribution of salaries between ```0``` and ```8000$``` / month.


````python
sns.displot(data = df[df["MonthlyIncome"] < 8000].MonthlyIncome, stat="density", kde=True)
````
