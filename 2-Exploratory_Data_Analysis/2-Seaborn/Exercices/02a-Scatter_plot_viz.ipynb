3. En créant un nuage de points, regardez comment se comporte la colonne _LotArea_ par rapport au _SalesPrices_

```python
sns.relplot(x="LotArea", y="SalePrice", data = dataset);
```

4. Refine your visualization by keeping only those houses that have a _LotArea_ of less than 20,000 square feet and a price of less than $500,000.

```python
sns.relplot(x="LotArea", y="SalePrice", data = dataset[(dataset["SalePrice"] < 500000) & (dataset["LotArea"] < 20000)]);
```

5. En créant un nuage de points, regardez la relation entre le _LotFrontage_ et le _LotArea_

```python
sns.relplot(x="LotFrontage", y="LotArea", data = dataset);
```

6. In the same way, refine your visualization by keeping only those houses that have a _LotFrontage_ of less than 200 square feet and a _LotArea_ of less than 100,000 square feet.


```python
sns.relplot(x="LotFrontage", y="LotArea", data = dataset[(dataset["LotArea"] < 100000) & (dataset["LotFrontage"] < 200)]);
```
