## Algebra


### Vector

For the rest of the course, it is important to understand what a scalar number is. For that, let's come back to what is a vector.


#### Definition

A vector is nothing more than a list of numbers. The size of this list can vary. For example, we can have :

### $\vec{u} = \begin{bmatrix}5\\2\end{bmatrix}$


This is a two dimensions vector, represented by $\vec{u}$ and can be represented like that: 

![](https://drive.google.com/uc?export=view&id=1upNvSWhsE04GiPbxt_udfb2wsFn1VKC_)


#### Operations on vectors

We can add two vectors between them. Either $\vec{u}$ and $\vec{v}$ two vectors with these coordinates:

### $\vec{u} = \begin{bmatrix}5\\2\end{bmatrix}$

### $\vec{v} = \begin{bmatrix}3\\4\end{bmatrix}$


### $\vec{u} + \vec{v} = \begin{bmatrix}5\\2\end{bmatrix} + \begin{bmatrix} 3\\4 \end{bmatrix} = \begin{bmatrix}8\\6\end{bmatrix}$







We can also multipicate a vector by a number. Now let's take $\vec{w}$  with these coordonates :

$\vec{w} = \begin{bmatrix}3\\2\end{bmatrix}$ 

$4 * \begin{bmatrix}3\\2\end{bmatrix}=\begin{bmatrix} 4 & * & 3\\ 4 & * & 2 \end{bmatrix}$
$= \begin{bmatrix}12\\8\end{bmatrix}$




The operation we just performed is: $4\vec{w}$ we can call a ** scalar number**


#### Vectors colinearity

Let's take three vectors $\vec{u}$, $\vec{v}$, $\vec{w}$ wich values are respectively :

## $\vec{u} = \begin{bmatrix}4\\2\end{bmatrix}$ 

## $\vec{v} = \begin{bmatrix}2\\6\end{bmatrix}$

## $\vec{w} = \begin{bmatrix}9\\7\end{bmatrix}$


We can express:

## $\vec{w} = 2\vec{u} + \frac{1}{2}\vec{v}$



That is to say that there is a linear relationship that allows to express $\vec{w}$ as a function of $\vec{u}$ and $\vec{v}.

In general, when you have two two-dimensional scalars, you can reach any point in a two-dimensional plane, EXCEPT if the vectors are collinear. That is to say that one can express $\vec{u}$ as a function of $\vec{v}$ or vice versa.

Let's take an example with two other vectors $\vec{i}$ and $\vec{j}$ respectively worth:

$\vec{i} = \begin{bmatrix}2\\4\end{bmatrix}$ 

$\vec{j} = \begin{bmatrix}1\\2\end{bmatrix}$




Here, we can see:

# $\vec{i} = 2\vec{j}$



In fact, we have a collinearity between the two vectors, which is going to be problematic because if these two vectors represented explanatory variables in a Machine Learning algorithm, it would mean that they are related to each other.

#### Matricial product

Very often we have to do matrix multiplications to be able to move in a plane. This is why it is good to know how matrix multiplications are made.

Let's take two matrices $X$ and $Y$ taking respectively the following values


### $X = \begin{bmatrix} 3 & 1 & 2\\ -1 & 3 & 2 \\ 3 & 1 & 4\end{bmatrix}$

### $Y = \begin{bmatrix} 2 & 1 & 0\\ -1 & 3 & 2 \\ 1 & 0 & 1\end{bmatrix}$



The product of the two matrices is produced as follows:

### $XY = \begin{bmatrix} 3 & 1 & 2\\ -1 & 3 & 2 \\ 3 & 1 & 4\end{bmatrix} \begin{bmatrix} 2 & 1 & 0\\ -1 & 3 & 2 \\ 1 & 0 & 1\end{bmatrix} = \begin{bmatrix} 6-1+2 & 3+3+0 & 0+2+2\\ -2-3+2 & -1+9+0 & 0+6+2 \\ 6-1+4 & 3+3+0 & 0+2+4\end{bmatrix} = \begin{bmatrix} 7 & 6 & 4 \\ -3 & 8 & 8 \\ 9 & 6 & 6 \end{bmatrix}$



More generally, matrix multiplication is done as follows:

## $\begin{bmatrix} a & b & c\\ d & e & f \\ g & h & i\end{bmatrix}$ $\begin{bmatrix} a' & b' & c'\\ d' & e' & f' \\ g' & h' & i'\end{bmatrix}$ $= \begin{bmatrix} aa'+ bd'+ cg' & ab'+be'+ch' & ac'+bf'+ci'\\ da'+ed'+fg' & db'+ee'+fh' & dc'+ef'+fi' \\ ga'+hd'+ig' & gb'+he'+ih' & ic'+if'+ii'\end{bmatrix}$



Take a minute to analyze how the multiplication is done on the first column of the matrix and you will easily understand the rest of the columns.


#### Matrix inversion

Knowing how to invert matrices will allow us to solve linear equations with several unknowns. In particular, this will be the basis for linear regressions in Machine Learning.

Consider the following equations:


## $2x + y + 4z = -1$


## $x + 2y - z = 0$


## $3x + y + 2z = 2$



This could be represented as follows:

### $\begin{bmatrix} 2&1&4\\1&2&-1 \\3&1&2 \end{bmatrix}$$\begin{bmatrix} x\\ y\\ z\end{bmatrix} = \begin{bmatrix} -1\\ 0\\ 2\end{bmatrix}$


Let's assign the first matrix to $A$ the second matrix to the vector $/vec{u}$ and the third to $\vec{v}$ such as :

### $A = \begin{bmatrix} 2&1&4\\ 1&2&-1 \\ 3&1&2 \end{bmatrix}$

### $\vec{u} = \begin{bmatrix} x \\ y \\ z \end{bmatrix}$

### $\vec{v} = \begin{bmatrix} -1 \\ 0 \\ 2 \end{bmatrix}$



As we had seen above, we can write: 

## $A\vec{u} = \vec{v}$

To resolve the equation, we can do:

## $\vec{u} = \frac{1}{A}\vec{v}$



When we are using matrices, we are not really writing $\frac{1}{A}$ but $A^{-1}$.

Here the result:

### $\vec{u} = A^{-1}\vec{v}$

This $A^{-1}$ represents the **inverse matrix** of $A$


We will not see in this course how to calculate the inverse of a matrix but it is important to understand this simple concept of solving equations which will be used later on to talk about Machine Learning.


## Probabilities ##

Let's move on to the probabilities. It is true that for some algorithms, we will need some theoretical knowledge of probability.


### Definition


#### Intuitive explanation

A probability is simply the "percentage chance" that an event will occur. This is not a rigorous definition but an intuitive one. A probability is therefore a number between 0 and 1.

For example, if we flip a coin, the probability of drawing is 50% or 0.5. Often the event "get tails" will be called A, the "get tails" event. Which will be modeled by:

### $P(A) = \frac{1}{2}$



#### What we should know on probabilities


##### What is independant events?

In probability, we try to describe events that are independent of each other. For example, if we again take the example of the coin toss. You can take events A "Get tails" and B "Get tails".

You will agree that it is impossible to get a coin flip if I flip the coin. This means that the events are independent.

Conversely, it is quite possible to have non-independent events. For example, let's take event A "having been drinking" and event B "having a car accident". It can be said that the probability of having a road accident is related to whether or not you have been drinking.

In general, <span style="text-decoration:underline;">we will always try to have independent events in an experiment. </span>


##### The sum of the independent event probabilities makes 1

If all the events in an experiment are independent then the sum of their probability is equal to 1.  

For example, if we restart the coin toss with A "Call heads" and B "Call tails" then

## $P(A) + P(B) = 1$



We also note the set of events independent of an experiment $\omega$ or, in fact,

## $P(\omega) = 1$




### Calculate of independant probabilities


#### Probabilities addition

When we add probabilities, we are trying to represent the probability that multiple events happen at the same time or separatly.

Let's take the example of a the 6 faces dice. The probability to throw the dice and fall on 1 or 6 is:

## $\frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}$



Mathematicly, we can write: 

Let A be the event "falling on 1" and B be the event "falling on 6."

## $P(A \cup B) = P(A) + P(B) = \frac{1}{3}$



#### Probabilities multiplication

Let's retake the example of the dice, but this time, we are not throwing une dice, but two.

We write the $A$ event: "Obtains 5 on the first dice" and $B$ event: "Obtains 5 on the second dice".

We can mathematicaly write: $A \cap B$ the event "Obtains 5 on the first and the second dice"

When **events are independants**, we can write:

## $P(A \cap B) = P(A) * P(B)$



In our example, the first throw of the dice is independant from the second one. We can say:

## $P(A \cap B) = P(A) * P(B) = \frac{1}{6} * \frac{1}{6} = \frac{1}{36}$



On the other hand, if we take dependent events then the formula differs. We'll have to do:

## $P(A \cap B) = P(A) + P(B) - P(A \cup B)$


It will be up to you to determine whether or not the events are dependent on experience.

#### Expected Value

Before we talk about the best known law of statistics, we need to explain what is the expectation as well as the variance.


##### If the random variable follows a discrete law

If the random variable follows a discrete law then the expectation is simply the mean :

## $E(X) = \frac{\sum_{0}^{n} x_{i} }{n}$


##### If the random varaible follow a continuous law

If our variable follow a continuous law then the expected value is: 

## $E(X) = \int_{-\infty}^{+\infty} xf(x)$



#### Variance

The variance is the difference between each point in your sample. The formula is written as follows:

## $V(X) = E(X - E(X))$



## Normal law


##### What is that?

There is a lot of talk about the normal law, but it is often difficult to understand why. Let's explain what it is first.

The normal distribution follows a Gaussian function:


## $f(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma })^{2}}$



The function is very complex but just imagine graphically that you get a bell curve.


##### The Central Theorem Limit

This law is very important thanks to the central limit theorem which says that, whatever the probability law that a random variable $X$ follows if we take an infinite number of samples, the mean of all these samples follows a normal law.  

So this means that it doesn't matter which initial law our random variable follows, since it can be considered to follow a normal law.
